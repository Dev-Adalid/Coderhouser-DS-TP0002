## Normalización de Variables

```python
from sklearn.preprocessing import MinMaxScaler

# Lista de columnas numéricas que quieres normalizar
columnas_numericas = ['blueTotalGold', 'redTotalGold', 'blueTotalExperience', 'redTotalExperience']

# Inicializar el scaler
scaler = MinMaxScaler()
```

```python
# Aplicar Min-Max Scaling a las columnas seleccionadas
df[columnas_numericas] = scaler.fit_transform(df[columnas_numericas])
```

## Codificación de Variables Categóricas:

```python
# Suponiendo que 'blueFirstBlood' es una variable categórica
df = pd.get_dummies(df, columns=['blueFirstBlood','redFirstBlood'], prefix=['blueFirstBlood','redFirstBlood'])
```

## Segmentacion por equipo a partir del dataset original
Aqui buscamos segmentar los resultados de partida por partida por equipo a fin de proveer mas registros en las pruebas individuales de alguos modelos solo dedicado a la performancia del equipo.

```python
# Segmentación por equipo a partir del dataset original
columnas = {
    "red": [item for item in list(df.columns.values) if "red" in item] + ["blueWins"],
    "blue": [item for item in list(df.columns.values) if "blue" in item],
}
```

```python
# Segmentación de equipos para previo análisis
blueTeam = df[columnas["blue"]]
# Supongamos que tienes un DataFrame llamado 'df' con columnas que tienen el prefijo 'blue'
new_column_names = [col.replace('blue', '') for col in columnas["blue"]]
# Asigna los nuevos nombres de columnas al DataFrame
blueTeam.columns = new_column_names

# se repite el proceso para el redTeam
redTeam = df[columnas["red"]]
new_column_names = [col.replace('red', '').replace('blue', '') for col in columnas["red"]]
redTeam.columns = new_column_names
```

```python
# invertimos la columna Wins la cual esta basada sobre el equipo Blue
redTeam["Wins"] = 1- redTeam["Wins"]
```

```python
# Guardar los DataFrames segmentados
blueTeam.to_csv( path_base + r"\\data\procesed\blueTeam.csv", index=False)
redTeam.to_csv( path_base + r"\\data\procesed\redTeam.csv", index=False)
```

```python
blueTeam["team"] = "Blue"
redTeam["team"] = "Red"
```

```python
# Concatenar los DataFrames para obtener un nuevo DataFrame combinado
data = pd.concat([blueTeam, redTeam])
data.to_csv( path_base + r"\\data\procesed\gameplay_by_team.csv", index=False)
```

## Análisis Exploratorio de Datos (EDA):


```python
# Descripción estadística para el total de registros de partidas
data.describe()
```

```python
# Reemplazar los valores de la columna "Wins"
data['WinTags'] = data['Wins'].replace({0: 'Loss', 1: 'Win'})

# Crear un mapeo de colores
color_map = {'Win': 'yellow', 'Loss': 'gray'}
```

```python
# Matriz de dispersión para variables seleccionadas sobre el total
features = ["Kills", "Assists", "Deaths", "WardsPlaced", "WardsDestroyed", "GoldDiff", "ExperienceDiff"]

fig = px.scatter_matrix(
    data,
    dimensions=features,
    color="WinTags",
    height=900,
    width=900,
    opacity=0.7,
    color_discrete_map=color_map
)

fig.update_traces(diagonal_visible=False)
fig.update_layout(title="Matriz de Dispersión para Variables Seleccionadas sobre el total")
fig.show()
fig.write_html(path_base + r"\\static\html\correlacion_total.html")
fig.write_image(path_base + r"\\static\img\correlacion_total.png")
```
