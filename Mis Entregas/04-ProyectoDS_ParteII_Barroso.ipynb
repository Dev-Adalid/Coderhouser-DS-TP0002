## Aplicacion de Moldelos

### Aplicando el modelo K-mean
Es importante analizar y explorar los clústeres para obtener información relevante. Por ejemplo, podríamos examinar si hay clústeres con un alto número de victorias (blueWins = 1) y qué características específicas tienen estas partidas en común. También podemos identificar clústeres con un alto número de derrotas (blueWins = 0) y analizar qué características los diferencian. Esta información podría proporcionar insights sobre las estrategias y tácticas exitosas o problemáticas que los equipos utilizan en las partidas.

```python
from sklearn.cluster import KMeans
```

```python
# Seleccionar las columnas relevantes para el modelo K-Means (características numéricas)
selected_features = df[['blueWardsPlaced', 'blueWardsDestroyed', 'blueKills', 'blueDeaths', 'blueAssists',
                        'blueEliteMonsters', 'blueDragons', 'blueHeralds', 'blueTowersDestroyed',
                        'blueTotalGold', 'blueAvgLevel', 'blueTotalExperience']]

# Normalizar las características para que tengan media 0 y varianza 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(selected_features)
```

```python
# Construir el modelo de K-Means
kmeans = KMeans(n_clusters=3, random_state=42)  # Supongamos que queremos 3 clústeres
kmeans.fit(scaled_features)

# Asignar las etiquetas de los clústeres a cada muestra
cluster_labels = kmeans.labels_

# Agregar los resultados del K-Means al DataFrame original
df['Cluster'] = cluster_labels
```

Al aplicar K-Means con 3 clústeres en el conjunto de datos, hemos agrupado las partidas clasificatorias en tres grupos distintos basados en las características numéricas seleccionadas. Cada clúster representa un conjunto de partidas similares en términos de las estadísticas proporcionadas en el dataset.

```python
# Crear una visualización para ver la distribución de los clústeres en el espacio de características
fig = px.scatter_3d(df, x='blueKills', y='blueTotalGold', z='blueTotalExperience', color='Cluster',
                    labels={'blueKills': 'Asesinatos', 'blueTotalGold': 'Oro Total', 'blueTotalExperience': 'Experiencia Total'},
                    title='Clústeres de Partidas Clasificatorias en el Espacio de Características',
                    color_continuous_scale=px.colors.sequential.Viridis)
fig.show()
```

Utilizando una visualización en 3D, hemos representado las partidas clasificatorias en el espacio de características seleccionado. Cada punto en el gráfico representa una partida, y los colores indican el clúster al que pertenece cada partida según los resultados del K-Means. Esto nos permite ver cómo se distribuyen las partidas en el espacio de características y cómo se agrupan en clústeres distintos.

### Aplicación de modelo PCA
Al aplicar PCA al conjunto de datos, hemos reducido la dimensionalidad de las características numéricas originales a un espacio de componentes principales. El número de componentes principales se establece previamente (en este caso, 3 componentes principales) y representa combinaciones lineales de las características originales.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Seleccionar las columnas relevantes para PCA (características numéricas)
selected_features = df[['blueWardsPlaced', 'blueWardsDestroyed', 'blueKills', 'blueDeaths', 'blueAssists',
                        'blueEliteMonsters', 'blueDragons', 'blueHeralds', 'blueTowersDestroyed',
                        'blueTotalGold', 'blueAvgLevel', 'blueTotalExperience']]

# Normalizar las características para que tengan media 0 y varianza 1
scaler = StandardScaler()
scaled_features = scaler.fit_transform(selected_features)

# Aplicar PCA
pca = PCA(n_components=3)  # Supongamos que queremos 3 componentes principales
principal_components = pca.fit_transform(scaled_features)
```

```python
# Crear DataFrame con los componentes principales
pca_df = pd.DataFrame(data=principal_components, columns=['Componente 1', 'Componente 2', 'Componente 3'])

# Agregar los resultados del PCA al DataFrame original
df_pca = pd.concat([df[['blueWins']], pca_df], axis=1)
```

Podemos analizar si hay una separación o agrupamiento claro entre las partidas ganadas (blueWins = 1) y las partidas perdidas (blueWins = 0) en el espacio de componentes principales. Esto puede indicar si existen patrones o tendencias distintivas en las características numéricas que se relacionan con el resultado del juego.

```python
# Crear una visualización para ver los datos en el espacio de componentes principales
fig = px.scatter_3d(df_pca, x='Componente 1', y='Componente 2', z='Componente 3', color='blueWins',
                    labels={'Componente 1': 'Componente Principal 1', 'Componente 2': 'Componente Principal 2', 'Componente 3': 'Componente Principal 3'},
                    title='Datos de Partidas Clasificatorias en el Espacio de Componentes Principales',
                    color_continuous_scale=px.colors.sequential.Viridis)
fig.show()
```

Utilizando una visualización en 3D, hemos representado las partidas clasificatorias en el espacio de componentes principales obtenido mediante PCA.
