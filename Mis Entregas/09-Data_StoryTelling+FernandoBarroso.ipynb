```python
# Importacion de Plotly como libreria de DataViz
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
historial = {
    "data": [
        {
            "x": [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
            "y": [5, 15, 32, 67, 92, 111, 128, 134, 140, 145, 150, 155],
            "name": 'Total de Usuarios',
            "type": 'bar'
        },
        {
            "x": [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
            "y": [1, 3, 6, 12, 18, 22, 25, 27, 28, 29, 30, 31],
            "name": 'Usuarios Conectados Diariamente',
            "type": 'lines',
            "line": {
                "dash": 'solid',
                "color": 'rgba(186,176,172,0.8)',
                "width": 5
            }
        }
    ]
}

fig = go.Figure()

for trace in historial['data']:
    if trace['type'] == 'bar':
        fig.add_trace(go.Bar(x=trace['x'], y=trace['y'], name=trace['name']))
    elif trace['type'] == 'lines':
        fig.add_trace(go.Scatter(x=trace['x'], y=trace['y'], mode='lines', name=trace['name'], line=trace['line']))

fig.update_layout(title='Crecimiento de Usuarios de League of Legends en Millones',
                  xaxis_title='Año',
                  yaxis_title='Usuarios',
                  legend=dict(
                    yanchor="top",
                    y=0.99,
                    xanchor="left",
                    x=0.01
                ))

fig.show()
print(path_base + r"\\static\html\Historial.html")
fig.write_html(path_base + r"\\static\html\Historial.html")
fig.write_image(path_base + r"\\static\img\Historial.png")
```

```python
droppedVariables = ["gameId", "blueFirstBlood", "blueEliteMonsters", "blueTotalGold", "blueTotalExperience", "blueTotalMinionsKilled", "blueTotalJungleMinionsKilled", "blueCSPerMin", "blueGoldPerMin", "redFirstBlood", "redKills", "redDeaths", "redAssists", "redEliteMonsters", "redTotalGold", "redTotalExperience", "redTotalMinionsKilled", "redTotalJungleMinionsKilled", "redGoldDiff", "redExperienceDiff", "redCSPerMin", "redGoldPerMin"]
dataCleaned = df.drop(droppedVariables, axis = 1)
```

```python
columnas_blue = [col for col in list(df.columns.values) if col.startswith("blue")]
blueTeam = df[columnas_blue]
blueTeam.columns = blueTeam.columns.str.replace(r'^blue|red', '', regex=True)
blueTeam["Team"] = "Blue"
columnas_red = [col for col in list(df.columns.values) if col.startswith("red")]+['blueWins']
redTeam = df[columnas_red]
redTeam.columns = redTeam.columns.str.replace(r'^blue|red', '', regex=True)
redTeam["Team"] = "Red"
```

```python
resultados = pd.concat([blueTeam,redTeam])
```

### Análisis Univariado

El siguiente gráfico superpuesto es útil para comparar visualmente cómo se distribuyen las colocaciones y destrucciones de guardianes de visión en un juego. Puedes observar si hay una superposición significativa en las distribuciones de ambas variables, lo que podría indicar una relación entre ellas. Por ejemplo, si las barras de "WardsDestroyed" son predominantemente más altas que las de "WardsPlaced", podría indicar una tendencia de destrucción activa de guardianes de visión en el juego. En resumen, este gráfico proporciona información sobre la dinámica de las estrategias de visión en el juego.

```python
# Crear un histograma superpuesto
fig = go.Figure()
fig.add_trace(go.Histogram(x=resultados['WardsPlaced'], name='WardsPlaced'))
fig.add_trace(go.Histogram(x=resultados['WardsDestroyed'], name='WardsDestroyed'))

# Overlay both histograms
fig.update_layout(barmode='overlay',title='Histograma Superpuesto de Wards')
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```

En este otro análisis podemos comprender la dinámica de los objetivos del juego relacionados con la derrota de dragones y heráldos. Puedes observar si existen similitudes en la frecuencia con la que se obtienen dragones y heráldos, o si hay diferencias notables en sus distribuciones. Esto puede proporcionar información sobre las estrategias utilizadas por los equipos en el juego y cómo influyen en el resultado.

```python
# Crear un histograma superpuesto
fig = go.Figure()
fig.add_trace(go.Histogram(x=resultados['Dragons'], name='Dragons'))
fig.add_trace(go.Histogram(x=resultados['Heralds'], name='Heralds'))

# Overlay both histograms
fig.update_layout(barmode='overlay',title='Histograma Superpuesto de Dragons y Heralds')
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```

### Análisis Bivariado

Este gráfico te permite visualizar cómo la cantidad de asesinatos ("Kills") se relaciona con el total de oro acumulado ("TotalGold"), considerando también el número de victorias ("Wins") utilizando colores. La línea de regresión lineal proporciona información sobre la tendencia general en los datos y si existe una relación significativa entre las variables. Puedes observar si, en promedio, un mayor número de asesinatos se traduce en un mayor total de oro y si esta relación afecta la probabilidad de ganar ("Wins").

```python
# Crear un gráfico de regresión lineal para analizar la relación entre "TotalGold" y "Kills"
fig = px.scatter(resultados, x="Kills", y="TotalGold", title="TotalGold vs. Kills (Regresión Lineal)", color="Wins", color_discrete_map="Team", trendline="ols")

# Mostrar el gráfico de regresión lineal
fig.show()
```

#### El análisis bivariado de estas columnas revela varias relaciones interesantes en los datos.

1. **Kills vs. Wins**: Existe una fuerte correlación positiva entre la cantidad de asesinatos (Kills) y las victorias (Wins). Esto sugiere que los equipos que obtienen más asesinatos tienden a ganar más partidas.
2. **Assists vs. Kills**: También se observa una correlación positiva entre las asistencias (Assists) y los asesinatos (Kills). Esto indica que los jugadores que realizan más asistencias suelen estar involucrados en más asesinatos.
3. **Deaths vs. Wins**: Se puede notar una correlación negativa entre las muertes (Deaths) y las victorias (Wins). Esto significa que equipos con menos muertes suelen tener más probabilidades de ganar.
4. **WardsPlaced y WardsDestroyed**: La cantidad de guardianes de visión colocados (WardsPlaced) está relacionada con la cantidad de guardianes de visión destruidos (WardsDestroyed). Esto sugiere que los equipos que colocan más guardianes de visión también tienden a destruir más de los guardianes del equipo contrario.
5. **GoldDiff y ExperienceDiff**: Las diferencias en oro (GoldDiff) y experiencia (ExperienceDiff) son indicadores críticos del rendimiento del equipo. Un mayor GoldDiff y ExperienceDiff generalmente se correlaciona con una mayor probabilidad de victoria. Esto significa que los equipos que obtienen ventajas significativas en términos de oro y experiencia tienden a ganar más partidas.

```python
plot_variables = ["Kills", "Assists", "Deaths", "WardsPlaced", "WardsDestroyed", "GoldDiff", "ExperienceDiff"]

fig = px.scatter_matrix(
    resultados,
    dimensions=plot_variables,
    color="Wins",
    height=900,
    width=900,
    opacity=0.7,
)

fig.update_traces(diagonal_visible=False)
fig.update_layout(title="Matriz de Dispersión para Variables Seleccionadas")
fig.show()
```
